```python
def encode(input_string):
    count = 1
    prev = ""
    lst = []
    for character in input_string:
        if character != prev:
            if prev:
                entry = (prev, count)
                lst.append(entry)
            count = 1
            prev = character
        else:
            count += 1
    entry = (character, count)
    lst.append(entry)
    return lst


def decode(lst):
    q = ""
    for character, count in lst:
        q += character * count
    return q
```

```python
class Solution:
    def longestCommonPrefix(self, strs: list[str]) -> str:
        res = "" #Taking a empty string for saving the result 
        for i in zip(*strs): # check out the bottom of this artical for its explanation.
            a = "".join(i) # joining the result of zip, check out the example
            if len(set(a)) != 1: # this will checkout, if the elements of the all the provided string on a same level (i.e a[i][i]) are identical or not.  If its identical then will always be equal to one.
                return res
            else:
                res += a[0] # if we are having identical element in the string will add it to our resulting string.

        return res
```

```python
s="abxabcxabcdx"
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        seen = ''
        max_length = 0

        for x in s:
            if x in seen:
                seen = seen[seen.index(x)+1:]

            seen+=x   
            max_length = max(max_length, len(seen))

        return max_length
    
    
sol=Solution()
sol.lengthOfLongestSubstring(s)
```

```python
class Node:
    def __init__(self,data=None) -> None:
        self.data=data
        self.next=None
        self.prev = None
         
class LinkedList:
    def __init__(self):
        self.head=Node()
        self.tail= None
        
    
    def append(self,data):
        new_node=Node(data)
        cur_node=self.head
        
        while cur_node.next != None:
            cur_node=cur_node.next
            
        new_node.prev=cur_node
        cur_node.next=new_node
        self.tail=new_node
        
        
    def length(self):
        count=0
        cur_node=self.head
        
        while cur_node.next != None:
            cur_node=cur_node.next
            count+=1
        return count
    
    
    def display(self):
        elems=[]
        cur_node=self.head
        while cur_node.next != None:
            cur_node=cur_node.next
            elems.append(cur_node.data)
        print(elems)
                
    def get_node(self,index):
        if index> self.length() or index<0:
            print("Error index out of range")
            return None
        step=0
        cur_node=self.head 
        while index > step:
            cur_node=cur_node.next
            step+=1
        print(f'node at index {index} is {cur_node.data}')
        return cur_node
    
    
    def reverse(self):
        elems=[] 
        last_node=self.get_node(self.length())
        while last_node.prev != None:
            elems.append(last_node.data)
            last_node=last_node.prev        
        return elems
    
    def reverse2(self):
       
        prev=None
        head=self.head
        
        while head:
            temp=head
            head=head.next
            temp.next=prev
            prev = temp
            
        return prev
            
            
        
        
        
        
        
    def insert(self,data,index):
        pass
    def erase(self,index):
        pass
        
        
        
        
```

```python
# insertion sort 
def insertion_sort(arr):
    
    for i in range(1,len(arr)):
        
        curr = arr[i]
        j = i 
        
        while j > 0 and arr[j-1] > curr:
            arr[j] = arr[j - 1]
            j-=1
        arr[j] = curr
        
    return arr

insertion_sort([2,3,0,4,1,0])
```

implement a function #stock_picker that takes in an array of stock prices,one for each hypothetical day. <br>
It should return a pair of days representing the best day to buy and the best day to sell. Days start at 0.<br>

> stock_picker([17,3,6,9,15,8,6,1,10])
=> [1,4]  # for a profit of $15 - $3 == $12


```python
def stock_poicker(stocks):
    left, right = 0 , 1
    prev  = stocks[0]
    curr = stocks[0] 
    results = [0,1]
    
    while right < len(stocks):
            
        if stocks[right] <= stocks[right-1]:
            left = right 
            curr = stocks[left]
            right+= 1 
        else:
            curr += stocks[right]
            right += 1 
            
        if curr > prev:
            results = [left, right-1]
            prev = curr
            
    return results    
        
        


```

```python
test = [17,3,1,9,10,14,15,8,6,1,10]

stock_poicker(test)
```

```python
def bubble_sort(arr):
    
    i=0
    j=0
    
    for i in range(len(arr)):
        swapped = False
        
        for j in range(len(arr)-i-1):
            
            if arr[j]> arr[j+1]:
                
                arr[j+1],arr[j]= arr[j],arr[j+1]
                swapped = True
            
            j+=1
        if not swapped:
            print("sorted on ", i+1)
            break
    return arr
```
